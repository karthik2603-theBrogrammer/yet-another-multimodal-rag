```python
import requests
import json
import base64
import os
def process_pdf(pdf_path, output_dir="pics"):
    url = 'http://localhost:8000/general/v0/general'
    
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    files = {
        'files': open(pdf_path, 'rb')
    }
    
    data = {
        'chunking_strategy': 'by_title',
        'strategy': 'hi_res',
        'extract_images_in_pdf': 'true',
        'infer_table_structure': 'true',
        'max_characters': '4000',
        'new_after_n_chars': '3800',
        'combine_text_under_n_chars': '2000',
        
        "extract_image_block_types": ["Image", "Table"],
        "extract_image_block_to_payload": 'true',
    }
    
    response = requests.post(
        url,
        files=files,
        data=data
    )
    
    files['files'].close()
    
    if response.status_code == 200:
        json_response = response.json()
        
        # Process and save images from the response
        for idx, element in enumerate(json_response):
            if 'image_base64' in element:
                # Decode and save image
                img_data = base64.b64decode(element['image_base64'])
                img_filename = f"{output_dir}/image_{idx}.png"
                with open(img_filename, 'wb') as img_file:
                    img_file.write(img_data)
                print(f"Saved image: {img_filename}")
            
            # Save tables if present
            if element.get('type') == 'Table' and 'image_base64' in element:
                table_filename = f"{output_dir}/table_{idx}.png"
                table_data = base64.b64decode(element['image_base64'])
                with open(table_filename, 'wb') as table_file:
                    table_file.write(table_data)
                print(f"Saved table: {table_filename}")
        
        # Save the full JSON response (optional)
        with open(f"{output_dir}/response.json", 'w') as f:
            json.dump(json_response, f, indent=2)
            
        return json_response
    else:
        print(f"Error: Status code {response.status_code}")
        print(response.text)
        return None

# Usage
pdf_path = 'split.pdf'
result = process_pdf(pdf_path)

# Print structure of elements
if result:
    for idx, element in enumerate(result):
        print(f"\nElement {idx}:")
        print(f"Type: {element.get('type', 'Unknown')}")
        print(f"Has image: {'image_base64' in element}")
        if 'text' in element:
            print(f"Text preview: {element['text'][:100]}...")
```

```python
import requests
import base64
import os
from typing import Dict, List, Any

def process_pdf(pdf_path: str, save_locally: bool = True) -> Dict[str, Any]:
    """
    Process PDF and get images in the response payload
    Args:
        pdf_path: Path to the PDF file
        save_locally: Whether to save images locally as well
    """
    url = 'http://localhost:8000/general/v0/general'
    
    files = {
        'files': open(pdf_path, 'rb')
    }
    
    data = {
        'chunking_strategy': 'by_title',
        'strategy': 'hi_res',

        'infer_table_structure': 'true',
        'max_characters': '4000',
        'new_after_n_chars': '3800',
        'combine_text_under_n_chars': '2000',
        # "extract_image_block_types": ["Image", "Table"],
        "extract_images_in_pdf": 'true',
        "extract_image_block_to_payload": 'true'  # This ensures images come back in response



    #         chunking_strategy="by_title",
    # max_characters=4000,
    # new_after_n_chars=3800,
    # combine_text_under_n_chars=2000,

    # infer_table_structure=True,
    # strategy='hi_res',

    # # extract_images_in_pdf=True,
    # # extract_image_block_output_dir='pics'

    # extract_image_block_types= ["Image", "Table"],
    # extract_image_block_to_payload=True
    }
    
    try:
        response = requests.post(
            url,
            files=files,
            data=data
        )
        
        files['files'].close()
        
        if response.status_code == 200:
            json_response = response.json()
            print(json_response)
            
            # Extract images from response
            images = []
            for idx, element in enumerate(json_response):
                if 'image_base64' in element:
                    image_data = {
                        'type': element.get('type', 'Unknown'),
                        'image': element['image_base64'],
                        'page_number': element.get('page_number'),
                        'coordinates': element.get('coordinates')
                    }
                    images.append(image_data)
                    
                    # Optionally save locally
                    if save_locally:
                        img_data = base64.b64decode(element['image_base64'])
                        os.makedirs('local_images', exist_ok=True)
                        img_filename = f"local_images/image_{idx}_{element.get('type', 'unknown')}.png"
                        with open(img_filename, 'wb') as img_file:
                            img_file.write(img_data)
                        print(f"Saved image locally: {img_filename}")
            
            return {
                'full_response': json_response,
                'images': images
            }
            
        else:
            print(f"Error: Status code {response.status_code}")
            print(response.text)
            return None
            
    except Exception as e:
        print(f"Error processing PDF: {str(e)}")
        return None

# Usage
pdf_path = 'pdfs/Karthik_Namboori_Intern_ATNT.pdf'
result = process_pdf(pdf_path)

if result:
    print(f"Number of images extracted: {len(result['images'])}")
    
    # Access individual images
    for idx, img_data in enumerate(result['images']):
        print(f"\nImage {idx}:")
        print(f"Type: {img_data['type']}")
        print(f"Page: {img_data['page_number']}")
        print(f"Coordinates: {img_data['coordinates']}")
        # The base64 image data is in img_data['image']
        
    # The full response with text and other elements is in result['full_response']
```
